```python
import cv2
import math
import numpy as np


image = cv2.imread("./image_2301540_20170629_ob_9266bb_angles-10.png")

pointangles = []

def anglepoints( event , x ,y ,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(image , (x,y) , 5 , (0,0,255) , cv2.FILLED)
        pointangles.append([x,y])
        if len(pointangles) != 1 :
            cv2.line(image ,tuple(pointangles[len(pointangles)-2]),(x,y),(0,0,255),2)
        print(pointangles)

    
def intercept(pt1,pt2):
    try :
        return (pt2[1]-pt1[1])/(pt2[0]-pt1[0])
    except :
        pass
    

def getAngle(points):
    point = points[-3:]
    a= points[0]
    b= points[1]
    c= points[2]
    m1 = intercept(b,a)
    m2 = intercept(b,c)
    radian = math.atan(abs((m2-m1)/(1 + (m2 * m1))))
    degreAngle = round(math.degrees(radian))
    
    return degreAngle

def calculate_angle(points):
    a = np.array(points[0]) # First
    b = np.array(points[1]) # Mid
    c = np.array(points[2]) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = round(np.abs(radians*180.0/np.pi))

    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
    
while True :
    if len(pointangles) % 3 == 0 and len(pointangles) != 0 :
        print("1 :",getAngle(pointangles))
        print("2 :" , calculate_angle(pointangles))
#     show the image 
    cv2.imshow("img",image)
#     draw circle on it cad do modifictaion 
    cv2.setMouseCallback("img",anglepoints)
#     repeat the boocle
    
    if cv2.waitKey(20) & 0xff == 27:
        pointangles = []
        image = cv2.imread("./image_2301540_20170629_ob_9266bb_angles-10.png")

#           break
            
cv2.destroyAllWindows()    
```
